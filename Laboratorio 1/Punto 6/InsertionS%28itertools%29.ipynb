##Insertion Sort - Permutations with itertools

```python
import itertools as iter
import matplotlib.pyplot as plt
```

```python
def insert(array):
    size = len(array)
    values = [0,0,1]
    key = 1
    entry = False
    while (key < size):
        i = key
        j = i - 1
        while (j > -1):
            values[1] = values[1]+1
            if array[i] < array[j]:
                entry = True
                values[2] = values[2] + 1
                aux = array[j]
                array[j] = array[i]
                array[i] = aux
            i = i - 1
            j = i - 1
        key = key + 1
    if entry == True:
        values[0] = (values[0]+6)*key+2
    else:
        values[0] = (values[0]+3)*key+2
    return [array,values]
```

```python

def generate(array,limit):
    stat = []
    lista = []
    swaps = []
    for e in iter.permutations(array,limit):
        lista = list(e)
        stat = insert(lista)
        swaps.append(stat[1][2])
    return swaps
```

```python
n = 8
array = []
values = []
for i in range(0,n):
    array.append(i)
values = generate(array,n)

plt.hist(values, bins='auto')
plt.title("Histograma - Numero de swaps para permutaciones cuando n = 9")
plt.show()
```
