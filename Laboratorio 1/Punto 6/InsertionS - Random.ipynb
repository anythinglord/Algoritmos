```python
import matplotlib.pyplot as plt
```

```python
swaps = []
def insert(array):
    size = len(array)
    values = [0,0,1]
    key = 1
    entry = False
    while (key < size):
        i = key
        j = i - 1
        while (j > -1):
            values[1] = values[1]+1
            if array[i] < array[j]:
                entry = True
                values[2] = values[2] + 1
                aux = array[j]
                array[j] = array[i]
                array[i] = aux
            i = i - 1
            j = i - 1
        key = key + 1
    if entry == True:
        values[0] = (values[0]+6)*key+2
    else:
        values[0] = (values[0]+3)*key+2
    return [array,values]
```

```python
def swap(array,i,j):
    t = array[i]
    array[i] = array[j]
    array[j] = t

def copia(array):
    v = []
    for i in range(0,len(array)):
        v.append(array[i])
    return v
```

```python
def perm(array,i,n):
    array2 = []
    if i == n:
        array2 = copia(array)
        stat = insert(array2)
        swaps.append(stat[1][2])
    else:
        for j in range(i,n):
            swap(array,i,j)
            perm(array,i+1,n)
            swap(array,i,j)
```

```python
n = 10
array = []
for i in range(0,n):
    array.append(i)
perm(array,0,n)
plt.hist(swaps, bins='auto')
plt.title("Histograma - Para el numero de Swaps")
plt.show()
```
