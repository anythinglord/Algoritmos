```python
import time as tm
import numpy as np
import matplotlib.pyplot as pl
```

```python
def fibn(n):
    t0 = tm.clock()
    a = 0
    b = 1
    for i in range(0,n):
        c = a + b
        b = a
        a = c
    time = tm.clock() - t0    
    return [b,time]
```

```python
numero = []
tiempo = []
for i in range(1,400):
    a = fibn(i)
    numero.append(i)
    tiempo.append(a[1])
```

```python
numero1 = np.array(numero)
tiempo1 = np.array(tiempo)
```

```python
x = pl.ylabel("Tiempo")
y = pl.xlabel('Numero')
pl.plot(numero1,tiempo1)
pl.show()
```
