```python
import itertools as iter
import matplotlib.pyplot as plt
```

```python
def bubble(S,n):
    swaps = 0
    for i in range(0,n-1):
        for j in range(0,n-i-1):
            if(S[j] > S[j+1]):
                swaps = swaps + 1
                temp = S[j]
                S[j] = S[j+1]
                S[j+1] = temp
    return swaps
```

```python

def generate(array,limit):
    stat = []
    lista = []
    swaps = []
    for e in iter.permutations(array,limit):
        lista = list(e)
        stat = bubble(lista,len(lista))
        swaps.append(stat)
    return swaps
```

```python
n = 8
array = []
values = []
for i in range(0,n):
    array.append(i)
values = generate(array,n)

plt.hist(values, bins='auto')
plt.title("Histograma - Numero de swaps para permutaciones cuando n = 9")
plt.show()
```
