```python
import matplotlib.pyplot as plt
```

```python
swaps=[]
def bubble(S,n):
    swaps = 0
    for i in range(0,n-1):
        for j in range(0,n-i-1):
            if(S[j] > S[j+1]):
                swaps = swaps + 1
                temp = S[j]
                S[j] = S[j+1]
                S[j+1] = temp
    return swaps
```

```python
def swap(array,i,j):
    t = array[i]
    array[i] = array[j]
    array[j] = t

def copia(array):
    v = []
    for i in range(0,len(array)):
        v.append(array[i])
    return v
```

```python
def perm(array,i,n):
    array2 = []
    if i == n:
        array2 = copia(array)
        stat = bubble(array2,len(array2))
        swaps.append(stat)
    else:
        for j in range(i,n):
            swap(array,i,j)
            perm(array,i+1,n)
            swap(array,i,j)
```

```python
n = 10
array = []
for i in range(0,n):
    array.append(i)
perm(array,0,n)
plt.hist(swaps, bins='auto')
plt.title("Histograma - Para el numero de Swaps")
plt.show()
```
